```python
# importing lib.
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#loading the image
im = cv2.imread("d.jpg")
```

```python
# resizing the image and covert its color format from bgr to gray
im2 = cv2.resize(im,(200,200))
im3 = cv2.cvtColor(im2,cv2.COLOR_BGR2GRAY)
```

```python
# function for ploting the image in gray color format
def draw(im):
    plt.imshow(im,cmap = 'gray')
    plt.style.use('seaborn')
    plt.axis("off")
    plt.title("imageshape" +str(im.shape))
    plt.show()
```

```python
# applying convolution on the image to make it blur or extraction the edges according ot the user choice
def convolution(img,filter_img):
    H = img.shape[1]
    W = img.shape[0]
    
    F = filter_img.shape[0]
    new_img = np.zeros((H-F+1,W-F+1))
    for row in range(W-F+1):
        for col in range(H-F+1):
            for i in range(F):
                for j in range(F):
                    new_img[row][col] += img[row+i][col+j]*filter_img[i][j]
                
            
                if new_img[row][col] > 255:
                    new_img[row][col] = 255
                elif new_img[row][col] < 0:
                    new_img[row][col] = 0
                
    return new_img



```

```python
# filter of 3*3 size used for blurring the image
blur_filter = np.ones((3,3))/9
```

```python
cimg = convolution(im3,blur_filter)
```
